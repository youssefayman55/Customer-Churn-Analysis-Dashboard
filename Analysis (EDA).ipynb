# 4- Data Visualization

```python
"""👉 This will tell you:
1- How many customers did you serve?
2- And how many are still working?  / 0 for still , 1 for churn
"""

sns.countplot(x= "Churn" , data =data )
plt.title("Churn Distribution" , color ="r" )
plt.xlabel("Churn" , color ="r")
plt.ylabel("Count" , color = "r")
plt.show()

# Insight : What we concluded from this is that more than 7000 customers will stay with us, while 3000 customers will leave and not continue.
```

```python
# 👉 Is there a difference between males and females in churn?

sns.countplot(x= "Gender" , hue="Churn" , data = data)
plt.title("Gender vs Churn" , color = "r")
plt.xlabel("Gender" , color = "r")
plt.ylabel("Count" , color = "r")
plt.show()

"""
Insight :
    ==> What we understand from this chart is that my clients are divided into males and females, 
    and more than 3500 females will continue with us, 
    while fewer than 1500 will leave. The same applies to males.
"""
```

```python
"""
👉 Key Insight:
    => Does a long contract reduce churn?
"""

sns.countplot(x= "Contract" , hue = "Churn" , data = data)
plt.title("Contract vs Churn" , color = "r" )
plt.xlabel("Contract" , color ="r")
plt.ylabel("Count" , color = "r")
plt.show()

"""
Insight :
==> What we understand from this chart is that payment period can affect customers,
    but in most cases, around 2500 customers have no problem with different payment period, 
    while less than 1000 customers might leave because of payment methods.
"""
```

```python
# 👉 Does a high price make people leave?

sns.boxenplot(x="Churn" , y ="MonthlyCharges" , data = data)
plt.title("Monthly Charges vs Churn" , color ="r")
plt.xlabel("Churn" , color ="r")
plt.ylabel("Monthly Charges" , color ="r")
plt.show()
```

```python
"""
👉 The most important question:
    =>Do new or old customers make more visits?
"""
sns.boxenplot(x="Churn" , y = "Tenure" , data =data)
plt.title("Tenure vs Churn" , color = "r")
plt.xlabel("Churn" , color ="r")
plt.ylabel("Tenure" , color = "r")
plt.show()
```

```python
"""
👉 This reveals:
    => The relationships between all the columns
"""
plt.figure(figsize=(8,5))
sns.heatmap(data.corr() , annot=True , cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()

"""
Insight :
    ==>What we will understand from this diagram is that there is a relationship between 
       (trunure and total charges) and (total charges and monthly charges)
"""
```

```python
sns.countplot(x="PaymentMethod", hue="Churn", data=data)
plt.title("Payment Method vs Churn" , color ="r")
plt.xlabel("Payment Method" , color ="r")
plt.ylabel("Count", color = 'r')
plt.xticks(rotation= 0)
plt.show()

"""
Insight :
 ==> What we understand from this graph is that there is a relationship between different payment methods 
     and whether customers continue to use the service or leave it. 
     The graph shows that more than 1750 customers have no problem with the payment methods and will continue to use the service, 
     while fewer than 750 businesses may have problems with those payment methods.
"""
```
